```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
from skimage.transform import resize
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

# Function to explore and visualize data distribution
def explore_data_distribution(data_dir, label_mapping, metadata):
    """
    Explore the dataset structure and visualize class distribution
    
    Args:
        data_dir: Directory containing the spectrograms
        label_mapping: Dictionary mapping primary_label to numerical label
        metadata: DataFrame containing metadata about classes
        
    Returns:
        DataFrame with counts of chirp files per class
    """
    chirp_count_per_class = defaultdict(int)
    file_type_counts = defaultdict(int)
    total_files = 0
    
    # Create reverse mapping for display
    label_to_class = {}
    for primary_label, label in label_mapping.items():
        class_name = metadata[metadata['primary_label'].astype(str) == primary_label]['class_name'].values
        if len(class_name) > 0:
            label_to_class[label] = class_name[0]
    
    # Count files
    for folder_name in os.listdir(data_dir):
        if folder_name in label_mapping:
            label = label_mapping[folder_name]
            class_name = label_to_class.get(label, f"Class {label}")
            folder_path = os.path.join(data_dir, folder_name)
            
            for filename in os.listdir(folder_path):
                if filename.endswith('.npy'):
                    total_files += 1
                    
                    # Track file types
                    if '_chirp.npy' in filename:
                        chirp_count_per_class[label] += 1
                        file_type_counts['chirp'] += 1
                    elif '_denoised.npy' in filename:
                        file_type_counts['denoised'] += 1
                    elif '_denoised_hpss.npy' in filename:
                        file_type_counts['denoised_hpss'] += 1
                    elif '_denoised_median_filter.npy' in filename:
                        file_type_counts['denoised_median_filter'] += 1
                    elif '_denoised_spectral.npy' in filename:
                        file_type_counts['denoised_spectral'] += 1
    
    # Convert to DataFrame for easier analysis
    chirp_df = pd.DataFrame([
        {'class_label': label, 'class_name': label_to_class.get(label, f"Class {label}"), 'chirp_count': count}
        for label, count in chirp_count_per_class.items()
    ]).sort_values(by='chirp_count', ascending=False)
    
    # Print statistics
    print("\n===== Dataset Exploration =====")
    print(f"Total number of .npy files: {total_files}")
    print("\nFile type distribution:")
    for file_type, count in file_type_counts.items():
        print(f"  {file_type}: {count} files")
    
    print("\n===== Class Distribution (chirp files) =====")
    print(f"{'Class Name':<30} | {'Count':<10}")
    print("-" * 43)
    
    for _, row in chirp_df.iterrows():
        print(f"{row['class_name']:<30} | {row['chirp_count']:<10}")
    
    # Create a bar chart of class distribution
    plt.figure(figsize=(12, 8))
    plt.bar(chirp_df['class_name'], chirp_df['chirp_count'])
    plt.xticks(rotation=90)
    plt.xlabel('Class')
    plt.ylabel('Number of Chirp Files')
    plt.title('Class Distribution - Number of Chirp Files')
    plt.tight_layout()
    plt.show()
    
    return chirp_df

def split_proportional_train_test(data_dir, label_mapping, metadata, target_shape=(128, 64), test_ratio=0.2):
    """
    Create a train/test split that preserves class proportions
    
    Args:
        data_dir: Directory containing the spectrograms
        label_mapping: Dictionary mapping primary_label to numerical label
        metadata: DataFrame containing metadata about classes
        target_shape: Target shape for resized spectrograms
        test_ratio: Proportion of data to use for testing
        
    Returns:
        X_train, y_train, train_files, X_test, y_test, test_files
    """
    # Step 1: Gather all chirp file paths by class label
    class_files = defaultdict(list)

    for folder_name in os.listdir(data_dir):
        if folder_name in label_mapping:
            label = label_mapping[folder_name]
            folder_path = os.path.join(data_dir, folder_name)
            for filename in os.listdir(folder_path):
                if '_chirp.npy' in filename:
                    file_path = os.path.join(folder_path, filename)
                    class_files[label].append(file_path)

    # Create a mapping from numerical label to class name
    label_to_class = {}
    for primary_label, label in label_mapping.items():
        class_name = metadata[metadata['primary_label'].astype(str) == primary_label]['class_name'].values
        if len(class_name) > 0:
            label_to_class[label] = class_name[0]
    
    # Step 2: Split while preserving proportions
    train_data, train_labels, train_files = [], [], []
    test_data, test_labels, test_files = [], [], []
    
    # For tracking distribution
    train_counts = defaultdict(int)
    test_counts = defaultdict(int)

    for label, files in class_files.items():
        # Sort for reproducibility
        files = sorted(files)
        
        # Split into train and test while preserving proportions
        class_train_files, class_test_files = train_test_split(
            files, test_size=test_ratio, random_state=42
        )
        
        # Load training samples
        for file_path in class_train_files:
            spectrogram = np.load(file_path)
            spectrogram = resize(spectrogram, target_shape, mode='reflect', anti_aliasing=True)
            spectrogram = np.expand_dims(spectrogram, axis=-1)  # Add channel dimension
            train_data.append(spectrogram)
            train_labels.append(label)
            train_files.append(file_path)
            train_counts[label] += 1

        # Load testing samples
        for file_path in class_test_files:
            spectrogram = np.load(file_path)
            spectrogram = resize(spectrogram, target_shape, mode='reflect', anti_aliasing=True)
            spectrogram = np.expand_dims(spectrogram, axis=-1)  # Add channel dimension
            test_data.append(spectrogram)
            test_labels.append(label)
            test_files.append(file_path)
            test_counts[label] += 1

    # Print distribution analysis
    print("\n===== Class Distribution After Proportional Split =====")
    print(f"{'Class Name':<30} | {'Training Count':<15} | {'Test Count':<15} | {'Total':<15} | {'Train %':<8}")
    print("-" * 90)
    
    total_train = 0
    total_test = 0
    
    # Print counts for each class
    for label in sorted(set(train_counts.keys()) | set(test_counts.keys())):
        class_name = label_to_class.get(label, f"Class {label}")
        train_count = train_counts[label]
        test_count = test_counts[label]
        total = train_count + test_count
        train_percentage = (train_count / total) * 100 if total > 0 else 0
        
        print(f"{class_name[:30]:<30} | {train_count:<15} | {test_count:<15} | {total:<15} | {train_percentage:.2f}%")
        total_train += train_count
        total_test += test_count
    
    print("-" * 90)
    overall_train_percentage = (total_train / (total_train + total_test)) * 100
    print(f"{'Total':<30} | {total_train:<15} | {total_test:<15} | {total_train + total_test:<15} | {overall_train_percentage:.2f}%")
    
    # Plot the distribution
    plt.figure(figsize=(12, 8))
    
    x = np.arange(len(label_to_class))
    width = 0.35
    
    # Convert to lists with the same order
    ordered_labels = sorted(list(set(train_counts.keys()) | set(test_counts.keys())))
    ordered_names = [label_to_class.get(label, f"Class {label}") for label in ordered_labels]
    ordered_train = [train_counts[label] for label in ordered_labels]
    ordered_test = [test_counts[label] for label in ordered_labels]
    
    plt.bar(x - width/2, ordered_train, width, label='Training')
    plt.bar(x + width/2, ordered_test, width, label='Testing')
    
    plt.xlabel('Class')
    plt.ylabel('Number of Files')
    plt.title('Class Distribution - Training vs Testing')
    plt.xticks(x, ordered_names, rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return (
        np.array(train_data), np.array(train_labels), train_files,
        np.array(test_data), np.array(test_labels), test_files
    )

# Function to load additional training data for each group
def load_additional_train_data(data_dir, label_mapping, file_type, train_chirp_files, target_shape=(128, 128)):
    """
    Load additional training data for a specific file type.
    Only loads files that correspond to the chirp files already in the training set.
    
    Args:
        data_dir: Directory containing the spectrograms
        label_mapping: Dictionary mapping primary_label to numerical label
        file_type: Type of spectrogram file to load (e.g., "_denoised.npy")
        train_chirp_files: List of chirp files in the training set
        target_shape: Target shape for resized spectrograms
        
    Returns:
        additional_data, additional_labels
    """
    additional_data = []
    additional_labels = []
    
    # Create a set of base filenames from training chirp files for efficient lookup
    base_filenames = {}
    for path in train_chirp_files:
        folder_name = os.path.basename(os.path.dirname(path))  # Get folder name (class ID)
        base_filename = os.path.basename(path).replace('_chirp.npy', '')  # Get base filename
        base_filenames[(folder_name, base_filename)] = path
    
    # Count matches found
    match_count = 0
    
    # Now load the corresponding files with the specified file_type
    for folder_name in os.listdir(data_dir):
        if folder_name in label_mapping:
            label = label_mapping[folder_name]
            folder_path = os.path.join(data_dir, folder_name)
            
            for filename in os.listdir(folder_path):
                if file_type in filename:
                    base_filename = filename.replace(file_type, '')
                    
                    # Check if we have the corresponding chirp file in training set
                    if (folder_name, base_filename) in base_filenames:
                        file_path = os.path.join(folder_path, filename)
                        try:
                            spectrogram = np.load(file_path)
                            spectrogram = resize(spectrogram, target_shape, mode='reflect', anti_aliasing=True)
                            spectrogram = np.expand_dims(spectrogram, axis=-1)
                            additional_data.append(spectrogram)
                            additional_labels.append(label)
                            match_count += 1
                        except Exception as e:
                            print(f"Error loading {file_path}: {e}")
    
    print(f"Found {match_count} matching {file_type} files for training chirp files")
    
    return np.array(additional_data), np.array(additional_labels)

# Function to print group distribution
def print_group_distribution(X_train, y_train, label_to_class, group_name):
    """
    Print the distribution of classes in a training group
    
    Args:
        X_train: Training features
        y_train: Training labels
        label_to_class: Dictionary mapping numerical label to class name
        group_name: Name of the group
    """
    class_counts = defaultdict(int)
    for label in y_train:
        class_counts[label] += 1
    
    print(f"\n===== Class Distribution for {group_name} =====")
    print(f"{'Class Name':<30} | {'Count':<10} | {'Percentage':<10}")
    print("-" * 55)
    
    total = len(y_train)
    
    for label in sorted(class_counts.keys()):
        class_name = label_to_class.get(label, f"Class {label}")
        count = class_counts[label]
        percentage = (count / total) * 100
        
        print(f"{class_name[:30]:<30} | {count:<10} | {percentage:.2f}%")
    
    print("-" * 55)
    print(f"{'Total':<30} | {total:<10} | 100.00%")
    
    # Plot the distribution
    plt.figure(figsize=(12, 8))
    
    # Convert to lists with the same order for plotting
    ordered_labels = sorted(class_counts.keys())
    ordered_names = [label_to_class.get(label, f"Class {label}") for label in ordered_labels]
    ordered_counts = [class_counts[label] for label in ordered_labels]
    
    plt.bar(ordered_names, ordered_counts)
    plt.xlabel('Class')
    plt.ylabel('Number of Files')
    plt.title(f'Class Distribution - {group_name}')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Function to evaluate model
def evaluate_model(model, history, X_test, y_test, le, group_name=None):
    """
    Evaluate the model and visualize results
    
    Args:
        model: Trained model
        history: Training history
        X_test: Test features
        y_test: Test labels (one-hot encoded)
        le: LabelEncoder used for class names
        group_name: Name of the group
        
    Returns:
        Dictionary with evaluation metrics
    """
    # 1. Plot train and validate accuracy / loss curve
    plt.figure(figsize=(12, 5))
    
    title_suffix = f" - {group_name}" if group_name else ""

    # Accuracy curve
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title(f'Training and Validation Accuracy{title_suffix}')

    # Loss curve
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title(f'Training and Validation Loss{title_suffix}')

    plt.tight_layout()
    plt.show()

    # 2. Model predict for report
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = np.argmax(y_test, axis=1)

    # 3. Output classification report
    print(f"\nClassification Report{title_suffix}:")
    print(classification_report(y_true, y_pred, target_names=le.classes_))

    # 4. Plot Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
    plt.figure(figsize=(10, 8))
    disp.plot(cmap='Blues', values_format='d')
    plt.title(f"Confusion Matrix{title_suffix}")
    plt.show()

    # 5. ROC Curve (multi-classification one-vs-rest)
    plt.figure(figsize=(12, 8))

    for i, class_name in enumerate(le.classes_):
        fpr, tpr, _ = roc_curve(y_test[:, i], y_pred_probs[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{class_name} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Multi-class ROC Curve (One-vs-Rest){title_suffix}')
    plt.legend(loc='lower right', fontsize='small')
    plt.grid(True)
    plt.show()
    
    # Return metrics for comparison
    return {
        'y_pred': y_pred,
        'y_true': y_true,
        'y_pred_probs': y_pred_probs,
        'accuracy': history.history['val_accuracy'][-1]
    }

# Function to create model
def create_model(input_shape, num_classes):
    """
    Create and compile a CNN model
    
    Args:
        input_shape: Shape of input data
        num_classes: Number of output classes
        
    Returns:
        Compiled model
    """
    input_layer = Input(shape=input_shape)
    
    # Convolutional layers
    x = Conv2D(32, (3, 3), activation='relu')(input_layer)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(128, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    
    # Fully connected layers
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dense(64, activation='relu')(x)
    output_layer = Dense(num_classes, activation='softmax')(x)
    
    # Create and compile model
    model = Model(input_layer, output_layer)
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Main function
def main():
    # Config
    data_dir = 'mel_spectrograms'
    csv_path = 'taxonomy.csv'
    target_shape = (128, 64)
    test_ratio = 0.2
    
    # Loading metadata
    metadata = pd.read_csv(csv_path)
    
    # Encode class_name to numerical label
    le = LabelEncoder()
    metadata['label'] = le.fit_transform(metadata['class_name'])
    
    # Map primary_label -> label
    label_mapping = dict(zip(metadata['primary_label'].astype(str), metadata['label']))
    
    # Map label -> class_name (for display)
    label_to_class = dict(zip(metadata['label'], metadata['class_name']))
    
    # Define the 4 groups
    groups = [
        {
            'name': 'Group 1: chirp + denoised',
            'file_types': ['_chirp.npy', '_denoised.npy']
        },
        {
            'name': 'Group 2: chirp + denoised_hpss',
            'file_types': ['_chirp.npy', '_denoised_hpss.npy']
        },
        {
            'name': 'Group 3: chirp + denoised_median_filter',
            'file_types': ['_chirp.npy', '_denoised_median_filter.npy']
        },
        {
            'name': 'Group 4: chirp + denoised_spectral',
            'file_types': ['_chirp.npy', '_denoised_spectral.npy']
        }
    ]
    
    # Lists to store results
    history_list = []
    model_list = []
    evaluation_results = []
    
    # Step 1: Explore data distribution
    print("\n" + "="*60)
    print("STEP 1: EXPLORING DATA DISTRIBUTION")
    print("="*60)
    chirp_df = explore_data_distribution(data_dir, label_mapping, metadata)
    
    print("\n" + "-"*60 + "\n")
    
    # Step 2: Get proportional train/test split for chirp files
    print("\n" + "="*60)
    print("STEP 2: CREATING PROPORTIONAL TRAIN/TEST SPLIT")
    print("="*60)
    X_train_chirp, y_train_chirp, train_files_chirp, X_test, y_test_raw, test_files = split_proportional_train_test(data_dir, label_mapping, metadata, target_shape=target_shape, test_ratio=test_ratio)
    
    # Convert test labels to categorical
    num_classes = len(np.unique(y_test_raw))
    y_test = to_categorical(y_test_raw, num_classes=num_classes)
    
    print(f"\nBase training set (chirp only): {len(X_train_chirp)} samples")
    print(f"Test set: {len(X_test)} samples")
    
    # Early stopping callback
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=5,
        min_delta=0.001,
        restore_best_weights=True,
        verbose=1
    )
    
    # Step 3: Train and evaluate models for each group
    print("\n" + "="*60)
    print("STEP 3: TRAINING AND EVALUATING MODELS")
    print("="*60)
    
    for group in groups:
        print(f"\n{'='*80}")
        print(f"Processing {group['name']}")
        print(f"{'='*80}")
        
        # Start with chirp files (already balanced)
        X_train_combined = X_train_chirp.copy()
        y_train_combined = y_train_chirp.copy()
        
        # Add the other file type from this group
        other_file_type = [ft for ft in group['file_types'] if ft != '_chirp.npy'][0]
        X_additional, y_additional = load_additional_train_data(
            data_dir, label_mapping, other_file_type, train_files_chirp, target_shape
        )
        
        # Combine datasets if we found additional files
        if len(X_additional) > 0:
            X_train_combined = np.concatenate([X_train_combined, X_additional])
            y_train_combined = np.concatenate([y_train_combined, y_additional])
        
        # Print distribution for this group
        print_group_distribution(X_train_combined, y_train_combined, label_to_class, group['name'])
        
        # Convert to categorical
        y_train = to_categorical(y_train_combined, num_classes=num_classes)
        
        # Create and compile model
        model = create_model(X_train_combined[0].shape, num_classes)
        print(model.summary())
        
        # Train the model with early stopping
        print(f"\nTraining with {len(X_train_combined)} samples, testing with {len(X_test)} samples")
        history = model.fit(
            X_train_combined, y_train,
            validation_data=(X_test, y_test),
            epochs=20,
            batch_size=32,
            verbose=1,
            callbacks=[early_stopping]
        )
        
        # Store model and history
        history_list.append(history)
        model_list.append(model)
        
        # Evaluate the model
        print(f"\nEvaluating {group['name']}...")
        eval_result = evaluate_model(model, history, X_test, y_test, le, group['name'])
        evaluation_results.append(eval_result)
    
    # Step 4: Final comparison of all models
    print("\n" + "="*60)
    print("STEP 4: MODEL PERFORMANCE COMPARISON")
    print("="*60)
    
    # Accuracy comparison table
    print("\nAccuracy Comparison:")
    print(f"{'Group':40} | {'Final Validation Accuracy':25}")
    print("-" * 70)
    for i, group in enumerate(groups):
        print(f"{group['name']:40} | {evaluation_results[i]['accuracy']:.6f}")
    
    # Plot accuracy comparison across all models
    plt.figure(figsize=(12, 6))
    
    # Training accuracy
    plt.subplot(1, 2, 1)
    for i, history in enumerate(history_list):
        plt.plot(history.history['accuracy'], label=f"Group {i+1}")
    plt.title('Training Accuracy Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.grid(True)
    
    # Validation accuracy
    plt.subplot(1, 2, 2)
    for i, history in enumerate(history_list):
        plt.plot(history.history['val_accuracy'], label=f"Group {i+1}")
    plt.title('Validation Accuracy Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Print stopping epoch for each model
    print("\nTraining Duration:")
    print(f"{'Group':40} | {'Epochs Trained':15} | {'Early Stopped':15}")
    print("-" * 75)
    for i, history in enumerate(history_list):
        epochs_trained = len(history.history['loss'])
        early_stopped = epochs_trained < 20  # If we trained fewer than max epochs
        print(f"{groups[i]['name']:40} | {epochs_trained:15d} | {'Yes' if early_stopped else 'No':15}")
    
    # Class-wise accuracy comparison
    print("\nClass-wise Accuracy Comparison:")
    class_accuracies = {}
    
    # Calculate per-class accuracy for each model
    for i, result in enumerate(evaluation_results):
        y_true = result['y_true']
        y_pred = result['y_pred']
        
        for class_idx in range(num_classes):
            class_name = le.classes_[class_idx]
            if class_name not in class_accuracies:
                class_accuracies[class_name] = []
                
            # Get indices for this class
            class_indices = np.where(y_true == class_idx)[0]
            if len(class_indices) > 0:
                correct = np.sum(y_pred[class_indices] == class_idx)
                accuracy = correct / len(class_indices)
            else:
                accuracy = float('nan')
                
            class_accuracies[class_name].append(accuracy)
    
    # Print class-wise accuracies - fixing the syntax error in the original code
    print(f"{'Class':30} | " + " | ".join([f"Group {i+1}" for i in range(len(groups))]))
    print("-" * (30 + 13 * len(groups)))
    
    for class_name, accuracies in class_accuracies.items():
        acc_strings = [f"{acc:.4f}" if not np.isnan(acc) else "N/A" for acc in accuracies]
        print(f"{class_name:30} | " + " | ".join([f"{acc_str:10}" for acc_str in acc_strings]))
    
    # Find where models disagree
    print("\nAnalyzing samples where models disagree...")
    disagreements = []
    
    for i in range(len(X_test)):
        predictions = [evaluation_results[j]['y_pred'][i] for j in range(len(groups))]
        if len(set(predictions)) > 1:  # Models have different predictions
            disagreements.append({
                'index': i,
                'file': test_files[i],
                'true_class': le.classes_[evaluation_results[0]['y_true'][i]],
                'predictions': [le.classes_[pred] for pred in predictions]
            })
    
    print(f"\nFound {len(disagreements)} samples (out of {len(X_test)}) where models disagree")
    
    if disagreements:
        print("\nSample disagreements (first 5):")
        for i, d in enumerate(disagreements[:5]):
            print(f"Sample {i+1}: {os.path.basename(d['file'])}")
            print(f"  True class: {d['true_class']}")
            for j, group in enumerate(groups):
                print(f"  {group['name']}: predicted '{d['predictions'][j]}'")
            print()

if __name__ == "__main__":
    main()
```
